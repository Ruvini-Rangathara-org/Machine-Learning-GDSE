```python
import pandas as pd

df_sales = pd.read_csv('04 Handling Missing Data using Python_Sales.csv')
df_sales
```

```python
df_sales.info()
```

```python
df_sales_nan = df_sales.dropna()
```

```python
df_sales_nan1 = df_sales.dropna(how='all')
df_sales_nan1
```

```python
df_sales_nan2 = df_sales.dropna(thresh=1)
df_sales_nan2
```

```python
df_sales_1 = df_sales.fillna(0)
df_sales_1
```

```python
df_sales_2 = df_sales.fillna({"Day_Temp":df_sales.Day_Temp.mean(),
                              "No_of_Customers":df_sales.No_of_Customers.median(),
                              "Product_Quality":'A',
                              })
```

```python
df_sales_2
```

```python
 # forward fill & backward fill

df_sales_3 = df_sales.fillna(method='ffill')
df_sales_3
```

```python
df_sales_4 = df_sales.fillna(method='bfill')
df_sales_4
```

```python
df_sales_5 = df_sales.fillna(method='ffill', axis='columns')
df_sales_5
```

```python
df_sales_6 = df_sales.fillna(method='ffill', limit=1)
df_sales_6
```

# Interpolation
### ( defualt linear interpolation )

```python
df_sales_7 = df_sales.interpolate()
df_sales_7
```

```python
# Step 1: Convert the 'Date' column to datetime
df_sales['Date'] = pd.to_datetime(df_sales['Date'])
```

```python
df_sales_8 = df_sales.set_index("Date")
df_sales_8
```

```python
df_sales_9 = df_sales_8.interpolate(method='time')
df_sales_9
```
